```python
# Read dataset_summary.csv into a pandas DataFrame
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
def read_jsonl_file(file_name: str) -> pd.DataFrame:
    jsonL_path = Path(__file__).with_name(file_name) if '__file__' in globals() else Path(file_name)
    print(f"Trying to read: {jsonL_path.resolve()}")
    df = pd.read_json(jsonL_path, lines=True)
    print('DataFrame shape:', df.shape)
    return df
```

```python
def read_csv_file(file_name: str) -> pd.DataFrame:
    csv_path = Path(__file__).with_name(file_name) if '__file__' in globals() else Path(file_name)
    print(f"Trying to read: {csv_path.resolve()}")
    df = pd.read_csv(csv_path)
    print('DataFrame shape:', df.shape)
    return df
```

```python
df_decompiled = read_csv_file(f"{os.getenv('PROJECT_ROOT_DIR')}/dataset/decompiled_syntax_errors.csv")
```

```python
df_selected_cases = read_jsonl_file(f"{os.getenv('PROJECT_ROOT_DIR')}/dataset/test_cases.jsonl")
```

```python
df_decompiled = df_decompiled[~df_decompiled['file_hash'].isin(df_selected_cases['file_hash'].values)]
```

```python
df_selected_cases_2 = read_jsonl_file(f"{os.getenv('PROJECT_ROOT_DIR')}/dataset/test_cases_2.jsonl")
```

```python
df_selected_cases_2 = df_selected_cases_2[~df_selected_cases_2['file_hash'].isin(df_selected_cases['file_hash'].values)]
```

```python
df_selected_cases_2 = df_selected_cases_2[~df_selected_cases_2['compiled_success'] == True]
```

```python
df_final_file_hashes = pd.concat([df_selected_cases['file_hash'], df_selected_cases_2['file_hash']])
```

```python
df_final_test_dataset = df_decompiled[df_decompiled['file_hash'].isin(df_final_file_hashes.values)]
```

```python
df_final_test_dataset.to_csv(f"{os.getenv('PROJECT_ROOT_DIR')}/dataset/final_test_dataset.csv", index=False)
```
