```python
dataframe = pd.read_csv(INPUT_DATASET_CSV)
```

```python
dataframe.shape
```

```python
dataframe = dataframe[~dataframe["error_description"].str.contains(r"\bimport\b", case=True, regex=True,  na=False)]
```

```python
dataframe.shape
```

```python
def extract_file_path(error_description: str):
    m = re.search(r'File\s+"([^"]+)"', error_description)
    return m.group(1) if m else None
```

```python
dataframe["extracted_file_path"] = dataframe["error_description"].apply(extract_file_path)
```

```python
dataframe["extracted_file_path"].iloc[0]
```

```python
def contains_decompyle_warning(file_path):
    try:
        with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
            for line in f:
                if "# WARNING: Decompyle incomplete" in line:
                    return True
        return False
    except Exception:
        return False
```

```python
dataframe["has_decompyle_warning"] = dataframe["extracted_file_path"].apply(contains_decompyle_warning)
```

```python
completed_decompylation = dataframe[dataframe["has_decompyle_warning"] == False]
```
