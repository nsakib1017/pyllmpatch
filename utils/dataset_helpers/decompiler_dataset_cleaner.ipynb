```python

# BASE_PYTHON_FILES = Path(os.getenv("BASE_DIR_PYTHON_FILES"))
# INPUT_DATASET_CSV = Path(os.getenv("PROJECT_ROOT_DIR")) / "dataset" / "pycdc" / os.getenv("BASE_DATASET_NAME")
PATH_TO_PYLINGUAL_DATASET = Path(os.getenv("PROJECT_ROOT_DIR")) / "dataset" / "pylingual"
PATH_TO_PYCDC_DATASET = Path(os.getenv("PROJECT_ROOT_DIR")) / "dataset" / "pycdc"
PATH_TO_DATASETS = Path(os.getenv("PROJECT_ROOT_DIR")) / "dataset"
OUTPUT_FILE = Path(os.getenv("PROJECT_ROOT_DIR")) / "dataset" / "pypi_pylingual_dataset_merged_pycdc_pylingual.csv"
```

```python
pylingual_df = pd.read_csv(PATH_TO_PYLINGUAL_DATASET / "pypi_pylingual_dataset_merged_pylingual.csv")
pycdc_df = pd.read_csv(PATH_TO_PYCDC_DATASET / "pypi_pylingual_dataset_merged_pycdc.csv")
```

```python
print(pylingual_df.columns, pycdc_df.columns)
```

```python
merged_df = pd.concat([pylingual_df, pycdc_df], ignore_index=True)
```

```python
merged_df = merged_df.sample(frac=1).reset_index(drop=True)
```

```python
duplicates = merged_df[
    merged_df.duplicated(
        subset=["file_hash", "bytecode_version", "dataset", "decompiler"],
        keep=False
    )
]
```
